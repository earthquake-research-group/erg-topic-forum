# Why Python for earthquake science?

- Reproducible workflows
- Fast numerical work
- Access to ObsPy / ML tools
- Simple plotting


## Basic data types

```python
integer = 3
floating = 3.14
text = "earthquake"
boolean = True

print(type(integer))
print(type(floating))
print(type(text))
print(type(boolean))
```

## Containers

```python
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_dict = {"station": "AU.MEL", "magnitude": 3.2}

print(my_dict["station"])
```

## Control flow

```python
magnitude = 2.7

if magnitude >= 3:
    print("Locate event")
else:
    print("Probably noise")
```

## Loops

```python
stations = ["MEL", "OTW", "GIP"]

for s in stations:
    print("Processing", s)
```

## NumPy

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

x[:5]
```

## Pandas

```python
import pandas as pd

df = pd.DataFrame({
    "station": ["MEL", "OTW", "GIP"],
    "magnitude": [2.1, 3.0, 1.8]
})

df
```

## Matplotlib

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Synthetic waveform")
plt.show()
```

## A tiny scientific task

```python
magnitudes = np.array([1.2, 2.4, 3.1, 2.8])

large = magnitudes[magnitudes >= 2.5]

large
```
